```python
import torch
from transformers import MusicgenForConditionalGeneration, MusicgenProcessor
import torchaudio
import gradio as gr
import tempfile

# Load the pre-trained MusicGen model
model = MusicgenForConditionalGeneration.from_pretrained("facebook/musicgen-small")
processor = MusicgenProcessor.from_pretrained("facebook/musicgen-small")

# Set to inference mode
model.eval()

# Check if GPU is available
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

def generate_music(text_prompt):
    # Preprocess the prompt
    inputs = processor(text=[text_prompt], return_tensors="pt").to(device)

    # Generate audio tokens
    with torch.no_grad():
        audio_values = model.generate(**inputs, max_new_tokens=1024)

    # Convert to CPU
    audio_values = audio_values[0].cpu()

    # Save as temporary WAV file
    with tempfile.NamedTemporaryFile(suffix=".wav", delete=False) as f:
        torchaudio.save(f.name, audio_values, sample_rate=model.config.audio_encoder.sampling_rate)
        return f.name  # Return path to WAV file

# Set up Gradio interface
interface = gr.Interface(
    fn=generate_music,
    inputs=gr.Textbox(lines=2, placeholder="Describe the mood or scene..."),
    outputs=gr.Audio(label="Generated Music"),
    title="🎼 Text to Music Generator",
    description="Enter a description like 'a calm sunset by the beach' and generate music using AI 🎧",
    allow_flagging="never"
)

# Launch the app
if __name__ == "__main__":
    interface.launch()
```
